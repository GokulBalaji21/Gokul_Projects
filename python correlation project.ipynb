```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,7)
```

```python
df = pd.read_csv(r'C:\Users\gokul\Desktop\movies.csv')
```

```python
df.head()
```

```python
# Looking for missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}, {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
# Changing data types of column

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.dtypes
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_smp = df.sort_values(by = ['gross'], ascending = False)
```

```python
# Removing duplicate

df = df.drop_duplicates()
```

```python
# Looking the correlation of data

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df_smp.head()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'green'}, line_kws={'color':'red'})
```

```python
df.corr(numeric_only = True, method='pearson') #types = Pearson, kendall, spearman
```

```python
visual = df.corr(numeric_only = True, method='pearson')
sns.heatmap(visual, annot=True)
plt.title('Movies correlation of numeric feature')
plt.xlabel('Movie details')
plt.ylabel('Movie details')
plt.show()
```

```python
df_smp.head()
```

```python
#Convert every unnumeric values into numeric

df_numeric = df_smp

for col in df_numeric.columns:
    if (df_numeric[col].dtype == 'object'):
        df_numeric[col] = df_numeric[col].astype('category')
        df_numeric[col] = df_numeric[col].cat.codes

df_numeric.head()
```

```python
df_corr = df_numeric.corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
plt.title('movie correlation')
plt.xlabel('movie details')
plt.ylabel('movie details')
plt.show()
```

```python
corr_unstack = df_corr.unstack()
corr_unstack
```

```python
sorted_values = corr_unstack.sort_values()
sorted_values
```

```python
high_corr = sorted_values[(sorted_values) > 0.5]
high_corr
```
